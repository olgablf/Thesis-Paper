```
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import ast
import time
import os
```

```
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}

def get_scraped_links(base_list): #, accepted_words):    
    scraped_links = []   
    for url in base_list:
        parent = 'https://' + url.split('/')[2]
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text)
        
        for link in soup.select('h3 a'):            
            try:
                l = link.get('href')
                scraped_links.append(l)            
            except:
                print('no url')

    scraped_links = list(set(scraped_links))
        
    print(len(scraped_links))    
    return scraped_links
num = 0
links = []
# arithmos selidwn sto search
for i in range(0,10):
    num = num + 1
    newpage = 'https://www.newsbeast.gr/tag/koronoios/page/'+ str(num)

    links.append(newpage)
    print(newpage)
    
urls = get_scraped_links(links) #, accepted_words)

def remove_non_ascii(text):
    return ''.join([i if ord(i) < 128 else ' ' for i in text])

list1=[]
# arithmos selidwn * arithmo arthrwn ana selida
for link in urls:
    try:
        #link = link.replace(u"'", u"")
        dic = {}
        print("Checking: ",link, "   \n")
        response = requests.get(link, headers=headers)
        soup = BeautifulSoup(response.text)
            
        if (response.status_code != requests.codes.ok):
            print("REQUEST GOT ERROR STATUS CODE! Aborting.....")
            break
        
        title = soup.find("h1")
    
        if (title):
            dic['title'] = title.text
        else:
            print("ARTICLE NO: ", link, " HAS NO TITLE")
            dic['title'] = "NO TITLE"
        
            
        date = soup.find('div',{'class':'article-top-meta-date'})
        if (date):
            dic['date'] = date.text
        else:
            print("ARTICLE NO: ", link, " HAS NO date")
            dic['date'] = "NO date"
        
        article =  soup.select('.article-content')
        for s in soup.select('script'):
            s.extract()
        article_text = [a.text for a in article]
        dic['body'] = ' '.join(article_text).replace(u'\xad', u' ').replace(u'\xa0', u' ').replace(u'\n', u'')
        
        dic['url'] = link  
    
        comments = soup.select('.comment-content')
        if (comments):
            comments_text = [a.text for a in comments]
            dic['comments'] = '$$'.join(comments_text).replace(u'\xad', u' ').replace(u'\xa0', u' ').replace(u'\n', u'')
        else:
            dic['comments'] = 'NO COMMENTS'        
        list1.append(dic)        
        time.sleep(3)
    except:
        print(link, ' sth bad ')
dfinal = pd.DataFrame(list1)
```
