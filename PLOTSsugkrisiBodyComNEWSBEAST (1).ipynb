```
df_body['Date'] = pd.to_datetime(df_body.Date,dayfirst = False)
df_body.set_index(df_body.Date,inplace=True)
df_body
```

```
df_body.sort_index (inplace = False)
df_body
```

```
df_comments['Date'] = pd.to_datetime(df_comments.date,dayfirst=True)
df_comments.set_index(df_comments.Date,inplace=True)
df_comments.sort_index (inplace = True)
df_comments
```

```
df_body.set_index('Date',inplace=False)
df_body.sort_index(inplace=True)
df_body
```

```
df_body = df_body[(df_body.index>='2020-03-01')&(df_body.index<='2020-11-23')]  
df_body
```

```

#df_comments.set_index('Date',inplace=True)
df_comments.sort_index(inplace=True)
df_comments
```

```
df_comments = df_comments[(df_comments.index>='2020-03-01')&(df_comments.index<='2020-11-23')]  
df_comments
```

```
#ta grouparw ana arthro gia na parw to meso oro apo ola ta comments ana arthro
df_anaarthro = df_comments.groupby('title').mean()
df_anaarthro
```

```

#enwnw to grouparismeno ana arthro me to df_body.  To koino tous stoixeio einai to title
#ta Anger_intensity_x mou dinoun to sentiment apo to df_anaarthro
#ta Anger_intensity_y mou dinoun ta sentiment apo to df_body
merged = df_anaarthro.merge(df_body, on='title')

merged['Date'] = pd.to_datetime(merged.date,dayfirst=True)
merged.set_index('Date',inplace=True)
merged.sort_index(inplace=True)
merged
```

```
merged.tail(100)
```

```
merged['Μήνας'] = pd.DatetimeIndex(merged['date']).month
merged
```

```
#comments
merged['Χαρά'] = merged['Joy_intensity_x']#+merged['Joy_intensity_y']
merged['Λύπη'] = merged['Anger_intensity_x'] + merged['Fear_intensity_x']+ merged['Sadness_intensity_x']
merged
```

```

plt.figure(figsize = (10,7))
plt.plot(merged['Χαρά'], label = 'Xαρά')
plt.plot(merged['Λύπη'], label = 'Λύπη')

plt.legend()
plt.title('Χαρά VS Λύπη')
plt.xlabel('Ημερομηνία', fontsize=14)
plt.ylabel('Βαθμολογίες', fontsize=14)
```

```
#arthra
merged['Φόβος'] = merged['Fear_intensity_y']#+merged['Joy_intensity_y']
merged['Θυμός'] = merged['Anger_intensity_y'] 
```

```

plt.figure(figsize = (10,7))
plt.plot(merged['Φόβος'], label = 'Φόβος')
plt.plot(merged['Θυμός'], label = 'Θυμός')

plt.legend()
plt.title('Φόβος VS Θυμός')
plt.xlabel('Ημερομηνία', fontsize=14)
plt.ylabel('Βαθμολογίες', fontsize=14)
```
