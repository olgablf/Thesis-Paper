```
df_anaarthro = df_comments.groupby('headline').mean()
df_anaarthro
```

```
merged = df_anaarthro.merge(df_body, on='headline')

merged['Date'] = pd.to_datetime(merged.date,dayfirst=True)
merged.set_index('Date',inplace=True)
merged.sort_index(inplace=True)
merged
```

```
plt.figure(figsize=(10,7))

#kinoumenos mesos 60 simeiwn gia to sentiment apo to arthro
plt.plot(merged.index,merged.Anger_intensity_x.rolling(60,center=True).mean(),label = 'Θυμός στα σχόλια')
#kinoumeno mesos 60 simeiwn gia to sentiment apo ta comments
plt.plot(merged.index,merged.Anger_intensity_y.rolling(60,center=True).mean(), label = 'Θυμός στα άρθρα ')

plt.legend()
plt.title('Θυμός σε σχόλια Vs Θυμός σε άρθρα (Καλοκαίρι)')
plt.xlabel('Ημερομηνία', fontsize=14)
plt.ylabel('Ένταση', fontsize=14)
```

```
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```
df_comments = pd.read_csv("/gdrive/My Drive/Colab Notebooks/FINALpthemasentiment_comments.csv", encoding="utf-8")
df_comments
```

```
#wordcloud comments
```

```
text = df_comments.comments[10000]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```
stopwords = set(STOPWORDS)
stopwords.update(["δεν", "κατά", "στην", "τη", "θα","που", "Aka", "την", "με","άλλες", "αυτή", "τις", "και", "μέσο", "σε", "του", "κάποτε", "των"])

# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white", min_font_size=6, width=800, height=400).generate(text)
# Display the generated image:
# the matplotlib way:
plt.figure( figsize=(10,7) )
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
