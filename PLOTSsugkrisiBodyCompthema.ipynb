```
df_comments = pd.read_csv("/gdrive/My Drive/Colab Notebooks/FINALpthemasentiment_comments.csv", encoding="utf-8")
df_comments
```

```
df_body['Date'] = pd.to_datetime(df_body['date'])
df_body
```

```
df_comments['Date'] = pd.to_datetime(df_comments.date)
df_comments
```

```

df_body.set_index('Date',inplace=True)
df_body.sort_index(inplace=True)
df_body
```

```

df_comments.set_index('Date',inplace=True)
df_comments.sort_index(inplace=True)

df_comments
```

```
df_anaarthro = df_comments.groupby('headline').mean()
df_anaarthro
```

```
merged = df_anaarthro.merge(df_body, on='headline')

merged['Date'] = pd.to_datetime(merged.date,dayfirst=True)
merged.set_index('Date',inplace=True)
merged.sort_index(inplace=True)
merged
```

```
plt.figure(figsize=(10,7))

#kinoumenos mesos 60 simeiwn gia to sentiment apo to arthro
plt.plot(merged.index,merged.Anger_intensity_x.rolling(60,center=True).mean(),label = 'Θυμός στα σχόλια')
#kinoumeno mesos 60 simeiwn gia to sentiment apo ta comments
plt.plot(merged.index,merged.Anger_intensity_y.rolling(60,center=True).mean(), label = 'Θυμός στα άρθρα ')

plt.legend()
plt.title('Θυμός σε σχόλια Vs Θυμός σε άρθρα')
plt.xlabel('Ημερομηνία', fontsize=14)
plt.ylabel('Ένταση', fontsize=14)
```
